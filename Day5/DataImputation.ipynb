##### _1/2- Ortalama/Orta Deger_

```python
from sklearn.impute import SimpleImputer
```

```python
imp=SimpleImputer(strategy='mean')

# Ortalama ile doldurma
```

```python
yenidata=imp.fit_transform(df)
```

```python
pd.DataFrame(yenidata) # değişkenler tırnak içerisinde yazılmaz.

# yenidatayı dataframe yaparak tablo halinde gözükmesini sağladık.
```

##### _3- En Yakın Komsu İle_

```python
from sklearn.impute import KNNImputer

# en yakın komsu ile doldurma
```

```python
knn=KNNImputer()
```

```python
dolu=knn.fit_transform(df)
```

```python
pd.DataFrame(dolu)
```

##### _4- Regression_

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
imp=IterativeImputer(estimator=lr)
lrdata=imp.fit_transform(df)
```

```python
pd.DataFrame(lrdata)
```
